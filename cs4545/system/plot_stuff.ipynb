```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_nodes = pd.read_csv("../../aggregated_output/transaction.csv", header=None)
df_nodes.columns = ['NumberNodes', 'Latency', 'TotalTransactions', 'TotalBytesSent']
print(df_nodes)
```

```python
fig, ax1 = plt.subplots()

ax1.plot(df_nodes['NumberNodes'], df_nodes['Latency'], color='orange', marker='o', label='Latency (seconds)')
ax1.set_xlabel('Number of Honest Nodes')
ax1.set_ylabel('Latency')
ax1.tick_params(axis='y')

ax2 = ax1.twinx()
ax2.plot(df_nodes['NumberNodes'], df_nodes['TotalMessages'], color='green', marker='s', label='Total Messages Sent')
ax2.set_ylabel('Total Messages')
ax2.tick_params(axis='y')

ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.title('Latency and Total Messages vs Number of Honest Nodes')
plt.show()
```

# Assignment 2 - Experiment 1

```python
df_a2_no_optimizations = pd.read_csv("../../aggregated_output/a2_no_optimizations.csv", header=None)
df_a2_no_optimizations.columns = ['NumberByzantine', 'NumberNodes', 'AverageLatency' , 'TotalMessages']
print(df_a2_no_optimizations)
```

```python
fig, ax1 = plt.subplots()


ax1.plot(df_a2_no_optimizations['NumberNodes'], df_a2_no_optimizations['AverageLatency'], color='orange', marker='o', label='Average Latency')
ax1.set_xlabel('Network Size (number of nodes)')
ax1.set_ylabel('Average Latency (seconds)')
ax1.tick_params(axis='y')


ax2 = ax1.twinx()
ax2.plot(df_a2_no_optimizations['NumberNodes'], df_a2_no_optimizations['TotalMessages'], color='green', marker='s', label='Total Messages Sent')
ax2.set_ylabel('Total Messages')
ax2.tick_params(axis='y')


ax1.legend(loc='upper left')
ax2.legend(loc='upper right')


plt.title('Latency and Total Messages vs Network Size Without Optimizations')
plt.show()
```

```python
df_a2_optimization_1 = pd.read_csv("../../aggregated_output/a2_optimization_1.csv", header=None)
df_a2_optimization_1.columns = ['NumberByzantine', 'NumberNodes', 'AverageLatency' , 'TotalMessages']
print(df_a2_optimization_1)
```

```python
fig, ax1 = plt.subplots()


ax1.plot(df_a2_optimization_1['NumberNodes'], df_a2_optimization_1['AverageLatency'], color='orange', marker='o', label='Average Latency ')
ax1.set_xlabel('Network Size (number of nodes)')
ax1.set_ylabel('Average Latency (seconds)')
ax1.tick_params(axis='y')


ax2 = ax1.twinx()
ax2.plot(df_a2_optimization_1['NumberNodes'], df_a2_optimization_1['TotalMessages'], color='green', marker='s', label='Total Messages Sent')
ax2.set_ylabel('Total Messages')
ax2.tick_params(axis='y')


ax1.legend(loc='upper left')
ax2.legend(loc='upper right')


plt.title('Latency and Total Messages vs Network Size With Optimization 1')
plt.show()
```

```python
# Create the plot
fig, ax1 = plt.subplots()

# Plot without optimizations
ax1.plot(df_a2_no_optimizations['NumberNodes'], df_a2_no_optimizations['AverageLatency'],
         color='orange', marker='o', linestyle='--', label='Latency (No Opt)', alpha=0.4)
ax1.plot(df_a2_optimization_1['NumberNodes'], df_a2_optimization_1['AverageLatency'],
         color='orange', marker='o', label='Latency (With Opt 1)')

ax1.set_xlabel('Network Size')
ax1.set_ylabel('Average Latency (seconds)')
ax1.tick_params(axis='y')

# Create a twin Y-axis for Total Messages
ax2 = ax1.twinx()

# Plot Total Messages
ax2.plot(df_a2_no_optimizations['NumberNodes'], df_a2_no_optimizations['TotalMessages'],
         color='green', marker='s', linestyle='--', label='Messages (No Opt)', alpha=0.4)
ax2.plot(df_a2_optimization_1['NumberNodes'], df_a2_optimization_1['TotalMessages'],
         color='green', marker='s',  label='Messages (With Opt 1)')

ax2.set_ylabel('Total Messages')
ax2.tick_params(axis='y')

# Add legends for both axes
ax1.legend(loc='best', ncol=2)
ax2.legend(loc='lower right', ncol=2)

# Add title and show the plot
plt.title('Latency and Total Messages vs Network Size (With and Without Optimization 1)')
plt.tight_layout()
plt.show()
```

```python
df_a2_no_optimizations.sort_values('NumberNodes', inplace=True)
df_a2_optimization_1.sort_values('NumberNodes', inplace=True)

df_comparison = pd.DataFrame({
    'NumberNodes': df_a2_no_optimizations['NumberNodes'],
    'Latency_No_Opt': df_a2_no_optimizations['AverageLatency'],
    'Latency_With_Opt': df_a2_optimization_1['AverageLatency'],
    'Messages_No_Opt': df_a2_no_optimizations['TotalMessages'],
    'Messages_With_Opt': df_a2_optimization_1['TotalMessages']
})

df_comparison['Latency_Percent_Decrease'] = (
    round((df_comparison['Latency_No_Opt'] - df_comparison['Latency_With_Opt']) / df_comparison['Latency_No_Opt'] * 100, 2)
)

df_comparison['Messages_Percent_Decrease'] = (
    round((df_comparison['Messages_No_Opt'] - df_comparison['Messages_With_Opt']) / df_comparison['Messages_No_Opt'] * 100, 2)
)

print(df_comparison[['NumberNodes', 'Latency_Percent_Decrease', 'Messages_Percent_Decrease']])
```
